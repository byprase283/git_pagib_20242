```
def tambahsampaike(n):
  jumlah = 0
  for i in range(n, 0, -1):
    jumlah += 1
    return jumlah
```

```
import time

waktu_mulai = time.perf_counter()
tambahsampaike(50000)
waktu_selesai = time.perf_counter()
print(f'lama waktu proses:{(waktu_selesai - waktu_mulai)/10000} detik')
```

```
#TSP(Traveling Salesman Problem)
import itertools

# Jarak antar Kota
distances = [
    [12,3,52,5],
    [90,9,2,3,],
    [34,4,65,2],
    [6,5,43,8]
]

n = len(distances)
cities = range(n)

min_path = None
min_cost = float('inf')

for perm in itertools.permutations(cities[1:]): #kata 0 sampai start
    path = (0,) + perm + (0,)
    cost = sum(distances[path[i]][path[i+1]] for i in range(len(path)-1))
    print(path)
    print(f'Biaya: {cost}')
    if cost < min_cost:
      min_cost = cost
      min_path = path

print(f'Rute Terbaik: {min_path}')
print(f'Biaya Minimum: {min_cost}')
```
